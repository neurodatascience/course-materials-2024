# Slow Python data structures

## Lists and loops

__List__ and __dictionaries__ are the main Python data structures. 

This is a list:

```python
numbers = [1, 2, 3, 4, 5]
```

Lists are accessed by index:

```python
numbers[2]
```

This is a dictionary:

```python
dictio = {"name": "Jin", "age": 42}
```

Dictionaries are accessed by key:

```python
dictio["name"]
```

Lists and dictionaries are often manipulated with for loops:

```python
def create_numbers(n_values):
    numbers = []
    for x in range(n_values):
        numbers += [1 + x]
    return numbers
```

```python
numbers = create_numbers(1000)
numbers
```

## Optimizing for loops

Python loops are slow:

```python
%%timeit
inverses = []
for x in numbers:
    inverses += [1 / x]
```

List comprehensions reduce the Python interpreter overhead:


```python
%%timeit
inverses_comp = [1 / x for x in numbers]
```

# Scipy
* Scientific computation library built upon Numpy
* Numpy provides basic functions while Scipy provides mathematical functions


**Differentiation**

```python
def f(x):
    return x**2 * np.sin(2 * x) * np.exp(-x)


x = np.linspace(0, 1, 100)

plt.plot(x, f(x));
```

```python
from scipy.misc import derivative

plt.plot(x, f(x))
plt.plot(x, derivative(f, x, dx=1e-5));
# plt.plot(x, derivative(f, x, dx=1e-5, n=2));
```

**Linear Algebra**

```python
from scipy.linalg import lstsq

# Sample dataset: house sizes (in square feet) and their prices (in dollars)
sizes = np.array([800, 1200, 1500, 2000, 2500])
prices = np.array([100000, 150000, 180000, 220000, 270000])

# We'll represent our data with the linear model: y = mx + b
# where y is the price, x is the size, m is the slope, and b is the intercept

# To use linear regression, we need to augment the features by adding a column of ones for the intercept term
X = np.column_stack((sizes, np.ones_like(sizes)))

# Use scipy's lstsq function to perform linear least squares regression
m, b = lstsq(X, prices)[0]

# Print the coefficients
print("Slope (m):", m)
print("Intercept (b):", b)

# Now, let's make predictions for new house sizes
new_sizes = np.array([1800, 2100])
predicted_prices = m * new_sizes + b
print("Predicted prices for new sizes:", predicted_prices)

# Now let's visualize our results
plt.scatter(sizes, prices, label="Given Prices")
plt.scatter(new_sizes, predicted_prices, label="Predicted Prices")
plt.xlabel("House size")
plt.ylabel("Prices")
plt.legend();
```

NumPy equivalent

```python
# Sample dataset: feature (X) and target variable (y)
X = np.array([800, 1200, 1500, 2000, 2500])  # Feature (e.g., house size)
y = np.array([100000, 150000, 180000, 220000, 270000])  # Target variable (e.g., house price)

# Add a column of ones for the intercept term
X_augmented = np.column_stack((X, np.ones_like(X)))

# Calculate the coefficients using the normal equation
theta = np.linalg.inv(X_augmented.T.dot(X_augmented)).dot(X_augmented.T).dot(y)

# Print the coefficients
slope, intercept = theta
print("Slope:", slope)
print("Intercept:", intercept)

# Make predictions for new feature values
new_X = np.array([1800, 2100])

new_X_augmented = np.column_stack((new_X, np.ones_like(new_X)))
predicted_y = new_X_augmented.dot(theta)
print("Predicted values for new feature values:", predicted_y)

# Now let's visualize our results
plt.scatter(sizes, prices, label="Given Prices")
plt.scatter(new_sizes, predicted_prices, label="Predicted Prices")
plt.xlabel("House size")
plt.ylabel("Prices")
plt.legend();
```

**Signal Processing**

```python
from scipy.signal import convolve2d

# Load up an image
image = np.load("../exercises/images.npz")

# Define a 3x3 smoothing kernel
kernel = np.ones((3, 3)) / 9

# Apply the kernel to the matrix using convolution
result = convolve2d(image["mni_template"], kernel, mode="same", boundary="fill", fillvalue=0)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].set_title(f"Image Size: {image['mni_template'].shape}")
axes[0].imshow(image["mni_template"], cmap="gray")

axes[1].set_title(f"Smoothed Image Size: {result.shape}")
axes[1].imshow(result, cmap="gray");
```
